#Etapa III

## Regressão linear

```python
# criar um script que separe em um novo data frame Production, de modo em que esse target vire tanto X como Y
```

```python
new_df = {'y':[]}
a = 0
b = 12
for nb in range(12):
   new_df[f'x{nb+1}'] = []
while b != df_total4['production'].shape[0]:
  var_x = df_total4['production'].iloc[a:b].values.reshape(-1).tolist()
  for nb in range(12):
    new_df[f'x{nb+1}'].append(var_x[nb])
  new_df['y'].append(df_total4['production'].iloc[b])
  a += 1
  b += 1
new_df=pd.DataFrame(new_df)
```

```python
new_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
X2 = new_df.drop(['y'], axis=1)
y2 = new_df['y']
```

```python
intervalo2 = int(new_df.shape[0]*.8)
```

```python
X_train2 = X2.iloc[:intervalo2, :]
X_test2 = X2.iloc[intervalo2:,:]
y_train2 = y2[:intervalo2]
y_test2 = y2[intervalo2:]
```

```python
model2 = LinearRegression()
model_t2 = model2.fit(X_train2, y_train2)
```

```python
predicao_model2 = model_t2.predict(X_test2)
mae = mean_absolute_error(predicao_model2, y_test2)
mae
```

```python
plt.figure(figsize=(25,7))
plt.plot(np.arange(y_test2.shape[0]),y_test2)
plt.plot(np.arange(y_test2.shape[0]),predicao_model2)
plt.show()
```

## Randon Forest com janela temporal

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

```python
nome_features = df_total4.columns
new_df_rf2 = {'y': []}
a = 0
b = 12
for name in nome_features:
    if name != 'production':  # Evitar criar listas para a coluna de produção
        for nb in range(10):
            new_df_rf2[f'{name}{nb+1}'] = []
while b < df_total4.shape[0]:
    for name in nome_features:
        if name != 'production':  # Evitar usar a coluna de produção para features deslocadas
            var_x = df_total4[name].iloc[a:b].values.tolist()
            for nb in range(10):
                new_df_rf2[f'{name}{nb+1}'].append(var_x[nb])
    new_df_rf2['y'].append(df_total4['production'].iloc[b])
    a += 1
    b += 1
new_df_rf2 = pd.DataFrame(new_df_rf2)
```

```python
new_df_rf2
```

```python
X3 = new_df_rf2.drop(['y'], axis=1)
y3 = new_df_rf2['y']
```

```python
intervalo = int(df_amostra.shape[0]*.8)
```

```python
X_train3 = X3.iloc[:intervalo,:]
X_test3 = X3.iloc[intervalo:,:]
y_train3 = y3[:intervalo]
y_test3 = y3[intervalo:]
```

```python
model3 = RandomForestRegressor(random_state = 42)
model_t3 = model3.fit(X_train3, y_train3)
```

```python
predicao_model3 = model_t3.predict(X_test3)
mae3 = mean_absolute_error(predicao_model3, y_test3)
mae3
```

```python
plt.figure(figsize=(25,7))
plt.plot(np.arange(y_test3.shape[0]),y_test3)
plt.plot(np.arange(y_test3.shape[0]),predicao_model3)
plt.show()
```
